```python
import pandas as pd
```

```python
original_data = pd.read_csv('../input/timesData.csv')
data = pd.read_csv('../input/timesData.csv')
```

```python
data.head()
```

```python
data["size_quality_rank"] = -1
```

```python
for i in range(0, len(data)):
    if(data['num_students'][i] is not None):
        if(isinstance(data['num_students'][i], str)):
            num_students = float(data['num_students'][i].replace(",",""))
        else:
            num_students = float(data['num_students'][i])
        if(data['citations'][i] is not None):
            citations = float(data['citations'][i])
            data["size_quality_rank"][i] = (10000000*citations) / ((num_students*1000)*(i+1))
```

```python
data = data.sort_values(by=['size_quality_rank'], ascending=False)
```
